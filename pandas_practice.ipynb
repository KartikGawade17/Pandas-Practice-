```python
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Amit', 'Rahul', 'Sneha', 'Neha']
})
```

```python
orders = pd.DataFrame({
    'customer_id': [1, 2, 2, 5],
    'order_amount': [500, 300, 700, 400]
})
```

```python
#Inner jOin = Which joins the records or columns present in both tables.
pd.merge(customers, orders, on='customer_id', how='inner')
```

```python
pd.merge(customers, orders, on='customer_id',how = 'left')
```

```python
pd.merge(customers, orders, on = 'customer_id', how = 'right')
```

```python
combine_data = pd.merge(customers, orders, on = 'customer_id', how = 'outer')
```

```python
combine_data
```

```python
combine_data['order_amount'] = combine_data['order_amount'].fillna(combine_data['order_amount'].mean())
```

```python
combine_data
```

```python
combine_data['name'] = combine_data['name'].fillna('Unknown')
```

```python
combine_data
```

```python
combine_data.dtypes
```

```python
combine_data['order_amount'] = combine_data['order_amount'].astype('int64')
```
