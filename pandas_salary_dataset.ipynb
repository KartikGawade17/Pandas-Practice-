```python
# Q.1 Find all employees who are located in Mumbai and show:Position, YearsExperience, Salary(INR)
data[data['Location']=="Mumbai"][['Position','YearsExperience']]
```

```python
# Q.2
data.groupby("EducationLevel")['Salary(INR)'].mean()
```

```python
#Q. 3
def salary_cat(salary):
    if salary > 1000000:
        return "High"
    elif 1000000 < salary > 600000:
        return "Medium"
    else:
        return "Low"

data['salary_category'] = data['Salary(INR)'].apply(salary_cat)
```

```python
data.head(5)
```

```python
# Q.4 Find employees working in IT industry with more than 5 years of experience, then sort them by Salary(INR) in descending order.
data[
    (data['Industry'] == "IT") & (data['YearsExperience'] > 5)
].sort_values(by= "Salary(INR)",ascending=False).head(5)
```

```python
#Q.5 Find how many employees are there in each Industry, and display the result sorted by count (highest first).
data.groupby('Industry')['Position'].count().sort_values(ascending=False)
```

```python
#Q.6 Find all employees whose Salary(INR) is between 700,000 and 1,500,000, and show:Position, Industry, Salary(INR)
data[data["Salary(INR)"].between(700000, 1500000)][["Position", "Industry", "Salary(INR)"]]
```

```python
#Q.7 Q2️⃣ (GroupBy + Multiple Aggregations) For each Industry, calculate:
#Average Salary
#Median Salary
#Maximum Salary
data.groupby("Industry")["Salary(INR)"].agg(['mean','median','max'])
```

```python
#Q.8
import numpy as np
data['Salary_Band'] = np.where(data['Salary(INR)'] >= 1500000, "Very High",
                               np.where(data['Salary(INR)'].between(1000000 , 1499999), "High", 
                                        np.where(data['Salary(INR)'].between(600000, 999999), "Medium", "Low")))
```

```python
#Q.9
data[(data["EducationLevel"].isin(values=['Masters']) | (data['EducationLevel'].isin(values=['PhD'])))]
```

```python
#Q.10 Find the number of employees per EducationLevel, then sort the result from highest count to lowest.
data.groupby("EducationLevel")['Position'].count().sort_values(ascending=False)
```
